```python
# Read the saved Excel file

df= pd.read_excel("myexcel.xlsx")
```

```python
# Display the dataframe

print(df)
```

```python
# 1. Determine the distribution of employees across each team and calculate the percentage split relative to the total number of employees.

team_count = df['Team'].value_counts()
team_percentage = (team_count / team_count.sum()) * 100

print(team_count)
print(team_percentage)
```

```python
# 2. Segregate employees based on their positions within the company.

position_count = df['Position'].value_counts()
print(position_count)
```

```python
#3.Identify the predominant age group among employees. 

age_bins = [18, 25, 35, 45, 55, 65]
age_labels = ['18-25', '26-35', '36-45', '46-55', '56-65']
df['Age Group'] = pd.cut(df['Age'], bins=age_bins, labels=age_labels)
common_age_group = df['Age Group'].value_counts().idxmax()

print(common_age_group)
```

```python
#4.Discover which team and position have the highest salary expenditure.

team_salary = df.groupby('Team')['Salary'].sum()
position_salary = df.groupby('Position')['Salary'].sum()

highest_team = team_salary.idxmax()
highest_position = position_salary.idxmax()

print(f"Highest Salary by Team: {highest_team}")
print(f"Highest Salary by Position: {highest_position}")
```

```python
#5. Investigate if there's any correlation between age and salary, and represent it visually. 

import seaborn as sns
import matplotlib.pyplot as plt

# Correlation coefficient
correlation = df['Age'].corr(df['Salary'])
print(f"Correlation between Age and Salary: {correlation}")

# Visualization
sns.scatterplot(data=df, x='Age', y='Salary')
plt.title('Correlation Between Age and Salary')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.show()
```

```python
#1. Distribution of Employees Across Teams

import matplotlib.pyplot as plt

# Bar Chart
team_percentage.plot(kind='bar', color='skyblue')
plt.title('Employee Distribution by Team (Percentage)')
plt.xlabel('Team')
plt.ylabel('Percentage')
plt.show()
```

```python
#2. Segregate Employees by Position

#Graph Type: Bar Chart

position_count.plot(kind='bar', color='orange')
plt.title('Employee Count by Position')
plt.xlabel('Position')
plt.ylabel('Count')
plt.show()
```

```python
# 3. Predominant Age Group

# Graph Type:Bar Chart

# Bar Chart for Age Groups
df['Age Group'].value_counts().sort_index().plot(kind='bar', color='green')
plt.title('Employee Distribution by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.show()
```

```python
# 4. Team and Position with Highest Salary Expenditure

# Graph Type: Bar Chart (Separate for Team and Position)

# Salary by Team
team_salary.sort_values().plot(kind='barh', color='purple')
plt.title('Total Salary by Team')
plt.xlabel('Total Salary')
plt.ylabel('Team')
plt.show()

# Salary by Position
position_salary.sort_values().plot(kind='bar', color='brown')
plt.title('Total Salary by Position')
plt.xlabel('Position')
plt.ylabel('Total Salary')
plt.show()
```

```python
# 5. Correlation Between Age and Salary

# Graph Type: Scatter Plot with Regression Line

import seaborn as sns

sns.regplot(data=df, x='Age', y='Salary', scatter_kws={'color': 'blue'}, line_kws={'color': 'red'})
plt.title('Correlation Between Age and Salary')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.show()
```
